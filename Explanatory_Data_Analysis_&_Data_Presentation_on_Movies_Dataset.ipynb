#<B>The best and the worst movies

```python
# importing HTML from ipython which will enable us to represent our data in high quality website like format
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best.head()
```

```python
# making 2 news columns in our dataframe
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
#checking top rows again
df_best.head()
```

```python
# setting columns for our  best movies dataframe
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
# setting Title as index
df_best.set_index("Title", inplace = True)
```

```python
#checking dataframe
df_best.head()
```

```python
# checking first row which contains link for poster image of our movies
df_best.iloc[0,0]
```

```python
# creating a subset of our df_best dataframe to demonstrate the poster images and how it look likes
subset = df_best.iloc[:5, :2]
subset
```

```python
# HTML code to return picture images
HTML(subset.to_html(escape=False))
```

```python
#sorting our Dataframe on average rating
df_best.sort_values(by = "Average Rating", ascending = False).head(4)
```

```python
# sorting our Dataframe on "ROI"
df_best.sort_values(by = "ROI", ascending = False).head(5)
```

```python
# Sorting on basis of budget
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending = False).head(5)
```

```python
# filling nan values in our df_best budget and votes columns
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
#checking info again
df_best.info()
```

```python
# defining a function 
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

##  Top 5 highest revenue

```python
best_worst(n = 5, by = "Revenue")
```

### **Top 5 Highest Budget**

```python
best_worst(5, "Budget")
```

### **Top 5 highest profit**



```python
best_worst(5, "Profit")
```

### **Top 5 - Lowest Profit**

```python
best_worst(5, "Profit", ascending = True)
```

## **Movies Top 5 - Highest ROI**

```python
best_worst(5, "ROI", min_bud = 50)
```

## **Movies Top 5 - Lowest ROI**

```python
best_worst(5, "ROI", ascending = True, min_bud = 100)
```

## **Movies Top 5 - Most Votes**

```python
best_worst(5, "Votes")
```

## **Movies Top 5 - Highest Rating**

```python
best_worst(5, "Average Rating", min_votes = 50)
```

## **Movies Top 5 - Lowest Rating**


```python
best_worst(5, "Average Rating", ascending = True, min_votes = 100)
```

```python
best_worst(5, "Average Rating", ascending = True, min_votes = 20, min_bud = 20)
```
